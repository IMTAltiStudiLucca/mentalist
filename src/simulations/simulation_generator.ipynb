```python
import pandas as pd
import yaml
```

```python
with open('attack_setup_base.yaml') as f:
    settings = yaml.safe_load(f)
```

```python
def to_file(r):
    networks={'CNN':'202101271243','NN':'202101100137'}
            
    settings['saved']['timestamp'] = networks[r.network_type]
    print(settings['saved']['timestamp'],r.network_type)
    settings['setup']['n_clients'] = r.n_clients
    settings['setup']['network_type'] = r.network_type
    settings['setup']['n_bits'] = r.n_bits
    settings['setup']['random_clients'] = r.random_clients
    filename = 'simulazione_{}.yaml'.format(r['Simuluation #'])
    with open(filename,'w') as fout:
        yaml.dump(settings, fout)
```

```python
simulations.apply(lambda x:to_file(x),axis=1)
```
